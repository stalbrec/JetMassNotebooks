```python
#https://matplotlib.org/ and google is you friend
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,12))
p = np.linspace(0,4)
ax.fill_between(p,5,1,facecolor='green',alpha=0.5, interpolate=True, label ="hello" )
ax.fill_between(p, p, 0,facecolor='blue',alpha=0.5, interpolate=True, label ="world" )
ax.plot(p,p,'r-',label='$f_1(P)=P$ ')
ax.set_xlabel("P [GeV]")
ax.set_ylabel("E/m")
ax.set_ylim(0,5)
ax.legend()
```
